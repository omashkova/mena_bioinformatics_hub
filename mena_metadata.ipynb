### mol_type

```python
unique, counts = np.unique(all_data.mol_type, return_counts=True)
c = list(counts)[0:3] + [counts[-1]] + [np.sum(counts[3:7])]
u = list(unique[0:3]) + [unique[-1]] + ['other']
```

```python
plt.figure(figsize=(12, 12))
plt.pie(c, labels=u)
plt.title('Nucleotide sequences, mol_type')
#plt.show()
plt.savefig('nucl_mol_type.png')
```

### countries

```python
unique, counts = np.unique(all_data.country_solo, return_counts=True)
```

```python
plt.figure(figsize=(12, 12))
plt.pie(counts, labels=unique)
plt.title('Nucleotide sequences, countries')
#plt.show()
plt.savefig('nucl_countries.png')
```

### tax_division

```python
unique, counts = np.unique(all_data.tax_division.dropna(), return_counts=True)
c = list(counts)[0:5] + list(counts)[7:10] + list(counts)[12:] + [np.sum(list(counts[5:7]) + list(counts[10:12]))]
u = list(unique)[0:5] + list(unique)[7:10] + list(unique)[12:] + ['other']
```

```python
plt.figure(figsize=(12, 12))
plt.pie(c, labels=u)
plt.title('Nucleotide sequences, tax_division')
#plt.show()
plt.savefig('nucl_tax_division.png')
```

## Metagenomes

```python
all_data = pd.read_csv('./Downloads/MENA_all_metagenomes_ENA - MENA_allmetagenomes_50k.csv')
all_data
```

```python
print(all_data.columns)
```

```python
print(np.unique(all_data.country))
```

```python
all_data['country_solo'] = [all_data['country'][i].split(' ')[0] for i in range(all_data.shape[0])]
all_data
```

```python
print(np.unique(all_data.country_solo))
```

```python
all_data['country_solo'] = [all_data['country_solo'][i].split(';')[0] for i in range(all_data.shape[0])]
all_data
```

```python
print(np.unique(all_data.country_solo))
```

```python
all_data['country_solo'] = [all_data['country_solo'][i] + ' Arab Emirates' if all_data['country_solo'][i] == 'United' 
                            else all_data['country_solo'][i] for i in range(all_data.shape[0])]
```

```python
print(np.unique(all_data.country_solo))
```

```python
all_data['country_solo'] = [all_data['country_solo'][i] + ' Arabia' if all_data['country_solo'][i] == 'Saudi' 
                            else all_data['country_solo'][i] for i in range(all_data.shape[0])]
```

```python
print(np.unique(all_data.country_solo))
```

```python
all_data['country_solo'] = [all_data['country_solo'][i].split(',')[0] for i in range(all_data.shape[0])]
all_data
```

```python
print(np.unique(all_data.country_solo))
```

### countries

```python
unique, counts = np.unique(all_data.country_solo, return_counts=True)
c = [list(counts)[0]] + list(counts)[3:5] + list(counts)[6:17] + [np.sum(list(counts[1:3])) + 27]
u = [list(unique)[0]] + list(unique)[3:5] + list(unique)[6:17] + ['other']
```

```python
plt.figure(figsize=(12, 12))
plt.pie(c, labels=u, autopct='%1.1f%%')
plt.title('Metagenomes, countries')
#plt.show()
plt.savefig('metagenome_countries.png')
```

### instrument_platform

```python
unique, counts = np.unique(all_data.instrument_platform, return_counts=True)
print(counts)
c = list(counts)[2:5] + [np.sum(list(counts[0:2])) + 33 + 257]
u = list(unique)[2:5] + ['other']
```

```python
plt.figure(figsize=(12, 12))
plt.pie(c, labels=u, autopct='%1.1f%%')
plt.title('Metagenomes, instrument_platform')
#plt.show()
plt.savefig('metagenome_instrument_platform.png')
```
